# Class with Constructor

```python
class Employee:
    
    # Constructor
    def __init__(self, salary=20000):
        self.salary = salary

    def computeBonus(self, bonus=0.05):
        return self.salary*bonus
```

```python
e3 = Employee(salary=100000)
print(e3.computeBonus())
print(e3.computeBonus(bonus=0.1))
```

# Datetime Data Type

```python
from datetime import datetime
```

```python
# Create Datetime
since_time = datetime(2020, 1, 2, 3, 4, 5)
print(since_time)
print(type(since_time))

since_date = datetime(2020, 1, 2)
print(since_date)
print(type(since_time))
```

```python
# Datetime Parameters
this_time = datetime.now()
print(type(this_time))
print('this time', this_time)
print('this day :', this_time.day)
print('this month :', this_time.month)
print('this year :', this_time.year)
```

```python
# Date Type
this_date = this_time.date()
print('this date :', this_date)
print('type :', type(this_date))
```

```python
# Time Operator
output_time = this_time + timedelta(hours=2)
print(output_time)

output_time = this_time + timedelta(days=2)
print(output_time)

# Date Operator
output_date = this_date + timedelta(days=2)
print(output_date)
```

```python
# Convert Datetime to String
date_str = str(since_date)
print(date_str)
print(type(date_str))

date_str = since_date.strftime('%Y-%m-%d %H:%M:%S')
print(date_str)
print(type(date_str))

date_str = since_date.strftime('%Y-%b-%d')
print(date_str)
print(type(date_str))
```

```python
# Convert String to Datetime
since_date = datetime.strptime('2020-01-01','%Y-%m-%d')
print(since_date)
print(type(since_date))

since_date = datetime.strptime('Jun 1 2005  1:33PM', '%b %d %Y %I:%M%p')
print(since_date)
print(type(since_date))
```
