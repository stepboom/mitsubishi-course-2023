# 2. An example chat API call

A chat API call has two required inputs:
- `prompt`: a text description of the desired image(s)
- `n`: the number of images to generate (between 1 to 10)
- `size`: the size of the generated images (should be one of 256x256, 512x512, or 1024x1024)
- `response_format`: the format in which the generated images are returned (url or b64_json)

```python
openai.api_key = 'INSERT_YOUR_API_KEY_HERE'
```

```python
# A text description of the desired image
PROMPT = "full view of mitsubishi ev car"
```

```python
# Example OpenAI Python library request
response = openai.Image.create(
    prompt=PROMPT,
    n=1,
    size="1024x1024",
    response_format="url",
)

response
```

As you can see, the response object has a few fields:
- `data`: a list of generated image objects (only one, unless you set `n` greater than 1)
    - `url`: the url of image generated by the model

Extract just the first image url with:

```python
url = response["data"][0]["url"]
print(url)
```

```python
# import library for reading and showing image

import matplotlib.pyplot as plt
from imageio import imread
```

```python
img = imread(url)
plt.imshow(img)
```
